```python
#import libraries
from bs4 import BeautifulSoup
import requests
```

```python
#connect to website and pull in data 
url = "https://www.airlinequality.com/airline-reviews/british-airways"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
soup.find_all('td')
```

```python
#lets try pulling various parts of review
review_boxes = soup.find_all('article', itemprop="review")

for i in review_boxes:
    try:
        review_title = i.find('h2', class_="text_header").text
        print("title:", review_title)
    except:
        print("null")
    try:
        review_body = i.find('div', class_="text_content").text
        print("review body:", review_body.split('|')[1])
    except:
        print("null")
    try:
        review_stats = i.find('table', class_ = 'review-ratings')
        print(review_stats)
        aircraft = review_stats.find('td', class_ = 'review-rating-header aircraft')['Aircraft']
        print(aircraft)
    except:
        print('null')
    break
```
